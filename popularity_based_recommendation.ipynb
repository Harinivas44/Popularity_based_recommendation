# **Criteria** For Popularity Based Recommendation Systems

The criteria is based on:
1. Movies with the highest rating
2. Number of views

```python
# where we can find the movies average rating
pop_df=pd.DataFrame(round(res_df.groupby('title')['rating'].mean(),1)).reset_index()
pop_df=pop_df.sort_values('rating',ascending=False)
pop_df.head()
```

```python
views=pd.DataFrame(res_df.groupby('title')['movieId'].count().reset_index())
views=views.sort_values('movieId',ascending=False)
views.head()
```

```python
popularity_df=pd.merge(pop_df,views,on='title')
popularity_df.head()
```

```python
# Here you fix the values you want to filter the movies based on above criteria.
views_count=int(input('Enter the no of views of the movies want to be:'))
ratings=float(input('Enter the ratings of the movies to filter out:'))
popularity_df[(popularity_df['rating']>=ratings) & (popularity_df['movieId']>=views_count)]
```
